```python
import re

'''Extract florida voters info'''
infoList = []
path = 'FloridaVoters.html'
with open(path, 'r') as florida:
    floridaList = florida.readlines()
    for eachline in floridaList:
        try:
            info = re.search('<td>([\w,-.\s]+)</td>', eachline).group(1)
            info = info.replace(',', '')
            infoList.append(info)
        except AttributeError:
            pass

# Combine each row into tuple
tupleList = []
for i in range(0, len(infoList), 6): # 6 columns 
    floridaTuple = (infoList[i], infoList[i + 1], infoList[i + 2])
    tupleList.append(floridaTuple)

# Sort tuple by the number of democratic voters  
sortedTuple = sorted(tupleList, key = lambda x: int(x[2]))
#print 'COUNTY', 'REPUBLICAN', 'DEMOCRAT'
for eachrow in sortedTuple:
    print eachrow[0], eachrow[1], eachrow[2]
```

```python
# Part(a)
fp = open("Quotes.txt",'r')
line = fp.readline()
text = []
while line is not '':
    line = line.rstrip()
    text.append(line)
    line = fp.readline()
fp.close()
quotes = text[0::2] 
spkrs = text[1::2]
full_quotes = []
for i in range (len(quotes)):
    full_quotes.append(quotes[i]+'  - '+spkrs[i])
full_quotes
```

```python
# Part(b)
import re
def to_word_list(full_quote):
    a = full_quote.lower()
    return re.split('\W+',a)    
```

```python
# Part (c)
def pl_dict(full_quotes): 
    pl_dict = {}
    for item in full_quotes:
        words = to_word_list(item)
        quote_dict = {}
        for w in words:
            if w in quote_dict:
                quote_dict[w] += 1
            else: 
                quote_dict[w] = 1
        pl_dict[item] = quote_dict
    return pl_dict 

#pl_dict(full_quotes)
```

```python
# Part(d)

def rev_pl_dict(full_quotes): 
    rev_pl_dict = {}
    all_word_list = []
    
    for item in full_quotes:
        words = to_word_list(item)
        for w in words:
            all_word_list.append(w)
            
    for w in all_word_list:
        quote_dict = {}
        for item in full_quotes:
            if w in to_word_list(item):
                quote_dict[item] = to_word_list(item).count(w)
            else:
                next 
        rev_pl_dict[w] = quote_dict

    return rev_pl_dict 

#rev_pl_dict(full_quotes)['country']
```

```python
# Part(e)
import math
def TF_IDF(word, quote):
    if word not in pl_dict(full_quotes)[quote]:
        tf_idf = 0 

    else:
        tf = float(pl_dict(full_quotes)[quote][word])/float(max(pl_dict(full_quotes)[quote].values()))
        idf = math.log(float(len(full_quotes)))/float(len(rev_pl_dict(full_quotes)[word].values()))
    tf_idf = tf*idf
    return tf_idf  
```

```python
# part(f)
def single_word_search(word):
    single_word_search = {}
    for quote in rev_pl_dict(full_quotes)[word]: 
        single_word_search[quote]=TF_IDF(word,quote)
    return single_word_search
```

```python
# part(g)
def quote_search_list_words(word_list):
    '''searches for all quotes where the words in the word list appear and returns the sum of tf-idf'''
    quote_search_all = []
    for word in word_list:
        quote_search = rev_pl_dict(full_quotes)[word].keys()
        for quote in quote_search:
            quote_search_all.append(quote)
    quote_search_all = list(set(quote_search_all))
    
    words_quote_math = {}
    for quote in quote_search_all:
        calculated_tf_idf = 0
        for word in word_list:
            calculated_tf_idf += TF_IDF(word, quote)
        words_quote_math[quote] = calculated_tf_idf
    return words_quote_math

quote_search_list_words(['sense', 'use'])
```
